```python
import numpy as np

from skimage import io
from skimage import data
#import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
from skimage import morphology
from matplotlib.pyplot import imshow,get_cmap 
import cv2 as cv2
import PIL
from PIL import Image, ImageDraw
```

Artère

```python
imageinitiale = cv2.imread("/content/drive/MyDrive/ROI1/patches artères/img7 A 1.png")
imagenoire = cv2.imread("/content/drive/MyDrive/ROI1/Black 150 584.png")
imageinitiale = cv2.cvtColor(imageinitiale, cv2.COLOR_RGB2GRAY)
imagenoire = cv2.cvtColor(imagenoire, cv2.COLOR_RGB2GRAY)
```

```python
f = 1
for i in range(imageinitiale.shape[0]):
  for j in range(imageinitiale.shape[1]):
    kernel = np.ones((f, f), np.uint8)
    res = cv2.erode(imageinitiale, kernel)
    if (res[i][j]!=imagenoire[i][j]).all():
      fig, ax = plt.subplots()
      ax.imshow(res, cmap=plt.cm.gray)
      f = f + 1 
      res = cv2.erode(imageinitiale, kernel)
print ("f = ",f-1)

```

Veine

```python
imageinitialev = cv2.imread("/content/drive/MyDrive/ROI1/patches veines/img8 V 1.png")
imagenoirev = cv2.imread("/content/drive/MyDrive/ROI1/Black 150 584.png")
imageinitialev = cv2.cvtColor(imageinitialev, cv2.COLOR_RGB2GRAY)
imagenoirev = cv2.cvtColor(imagenoirev, cv2.COLOR_RGB2GRAY)
```

```python
q = 1
for i in range(imageinitialev.shape[0]):
  for j in range(imageinitialev.shape[1]):
    kernel = np.ones((q, q), np.uint8)
    resv = cv2.erode(imageinitialev, kernel)
    if (resv[i][j]!=imagenoirev[i][j]).all():
      fig, ax = plt.subplots()
      ax.imshow(resv, cmap=plt.cm.gray)
      q = q + 1 
      resv = cv2.erode(imageinitialev, kernel)
print ("q = ",q-1)
```
