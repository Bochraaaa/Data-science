```python
import PIL
from PIL import Image

filename = "C:/Users/pc-hp/Desktop/New folder/camvid--FINAL/CamV/train/s_hc01_10_1.png"
img = Image.open(filename)
pixels = img.load()


#color the top sides (outside space of the layers) into GRAY
for i in range (0, 512):
    for j in range (0, 512):
        if img.getpixel((i,j)) != (255, 255, 255):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (105,105,105)
        else: 
            break
            
            
#color the bottum sides (outside space of the layers) into GRAY
for i in range (511, 0, -1):
    for j in range (511, 0, -1):
        if img.getpixel((i,j)) != (255, 255, 255):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (105,105,105)
        else: 
            break

            
            
#color the 1st black layer
for i in range (0, 512):
    for j in range (0, 512):
        if img.getpixel((i,j)) == (0, 0, 0):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (63, 72, 204)
            if img.getpixel((i,j+1)) == (255, 255, 255):
                break

                
                
#color the 2 black layer
for i in range (0, 512):
    for j in range (0, 512):
        if img.getpixel((i,j)) == (0, 0, 0):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (64, 192, 0)
            if img.getpixel((i,j+1)) == (255, 255, 255):
                break
                
                
                
#color the 3rd black layer
for i in range (0, 512):
    for j in range (0, 512):
        if img.getpixel((i,j)) == (0, 0, 0):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (192, 0, 128)
            if img.getpixel((i,j+1)) == (255, 255, 255):
                break
                
                
                
#color the 4th black layer                              
for i in range (0, 512):
    for j in range (0, 512):
        if img.getpixel((i,j)) == (0, 0, 0):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (128, 64, 128)
            if img.getpixel((i,j+1)) == (255, 255, 255):
                break 
                
                
#color the 5th black layer                
for i in range (0, 512):
    for j in range (0, 512):
        if img.getpixel((i,j)) == (0, 0, 0):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (0, 128, 64)
            if img.getpixel((i,j+1)) == (255, 255, 255):
                break 
                
                
                
#color the 6th black layer                
for i in range (0, 512):
    for j in range (0, 512):
        if img.getpixel((i,j)) == (0, 0, 0):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (128, 0, 0)
            if img.getpixel((i,j+1)) == (255, 255, 255):
                break 
                
                
#color the 7th black layer                
for i in range (0, 512):
    for j in range (0, 512):
        if img.getpixel((i,j)) == (0, 0, 0):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (64, 0, 128)
            if img.getpixel((i,j+1)) == (255, 255, 255):
                break  
                
                
#color the 8th black layer                
for i in range (0, 512):
    for j in range (0, 512):
        if img.getpixel((i,j)) == (0, 0, 0):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (255, 242, 0)
            if img.getpixel((i,j+1)) == (255, 255, 255):
                break 
                
                
# RE-#color the top and bottum sides (outside space of the layers) into BLACK - again
for i in range (0, 512):
    for j in range (0, 512):
        if img.getpixel((i,j)) != (255, 255, 255):
            #img.getpixel((i,j)) = (105,105,105)
            pixels[i,j] = (0, 0, 0)
        else: 
            break
for i in range (511, 0, -1):
    for j in range (511, 0, -1):
        if img.getpixel((i,j)) != (255, 255, 255):
            #img.getpixel((i,j)) = (105,105,105) v 
            pixels[i,j] = (0, 0, 0)
        else: 
            break

img.save('99999.png')
```

```python
#connect brocken lines using a horizontal kernel
import cv2, numpy as np
import PIL
import glob
import os
from PIL import Image
#bw = cv2.imread("12.png")
for filename in glob.glob('C:/Users/pc-hp/Desktop/data Augmentation/final data augmentation/XX/M/*.png'): 
    bw = cv2.imread(filename, 0)
    kernel = np.ones((2,15), np.uint8)  # note this is a horizontal kernel
    d_im = cv2.dilate(bw, kernel, iterations=1)
    e_im = cv2.erode(d_im, kernel, iterations=1)

    status = cv2.imwrite(filename, e_im)
    print("Image written to file-system : ",status)
    
```
