# Closing

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("/content/drive/MyDrive/mémoire/testing model ODIR/2992-2000/675_right.jpg", 0)


kernel = np.ones((3,3),np.uint8)   # 3x3 kernel with all ones. 



#Closing is opposit, dilation followed by erosion.
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
closing = cv2.cvtColor(closing,cv2.COLOR_GRAY2RGB)
cv2.imwrite('/content/drive/MyDrive/mémoire/675closing.jpg',closing)

fig, ax = plt.subplots()
ax.imshow(closing, cmap=plt.cm.gray)
```

Black-hat

```python
# Importing OpenCV and numpy
import cv2
# Defining the kernel to be used in Top-Hat
filterSize =(3, 3)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,
filterSize)
# Reading the image named 'input.jpg'
input_image = cv2.imread("/content/drive/MyDrive/mémoire/testing model ODIR/2992-2000/675_right.jpg")
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
# Applying the Black-Hat operation
tophat_img = cv2.morphologyEx(input_image,
cv2.MORPH_BLACKHAT,
kernel)


fig, ax = plt.subplots()
ax.imshow(tophat_img, cmap=plt.cm.gray)
cv2.imwrite('/content/drive/MyDrive/mémoire/675blackhat.jpg',tophat_img)
  
```

Clahe

```python
import cv2
import matplotlib.pyplot as plt
path = "/content/drive/MyDrive/mémoire/dataset/ODIR-5K/1024_left.jpg"
img = cv2.imread(path) 
#clahe
clahe = cv2.createCLAHE(clipLimit=3., tileGridSize=(1,1))
#lab = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)  # convert from BGR to LAB color space
l, a, b = cv2.split(lab)  # split on 3 different channels
l2 = clahe.apply(l)  # apply CLAHE to the L-channel
lab = cv2.merge((l2,a,b))  # merge channels
img2 = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)  # convert from LAB to BGR
#fig, ax = plt.subplots()
#ax.imshow(img2, cmap=plt.cm.gray)
cv2.imwrite('/content/drive/MyDrive/mémoire/j.jpg',img2)


```
