```python

import cv2
import PIL
from PIL import Image
```

```python
image = cv2.imread("/content/drive/MyDrive/fond d'oeil drive/AV/artères/ROI/ROI 02.png")
grayscale = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
cv2.imwrite("/content/drive/MyDrive/fond d'oeil drive/AV/artères/ROI/ROI 02 grayscale.png",grayscale)
```

```python
b = 0
b = b + 1
a = 1
img = Image.open("/content/drive/MyDrive/fond d'oeil drive/AV/grayscale/02.png")
w, h = img.size
x = 0
while x < h :

  y = 0
  while y < w :

    imm = img.crop(box=(x, y, x+100, y+100))
    imm.save("imgdrive" +str(b) + "_cropp" +str (a)+".png")
    a = a + 1
    y = y + 25
  x = x + 25
```
